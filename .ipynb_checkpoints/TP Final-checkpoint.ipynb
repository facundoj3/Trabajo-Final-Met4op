```python
import pandas as pd
import geopandas as gpd
import numpy as np
import seaborn as sns
import matplotlib.pylab as plt
import json
from shapely.geometry import shape
from adjustText import adjust_text
from matplotlib.colors import Normalize
```

```python
def split_str_into_lines(input_str: str, line_length: int):
    words = input_str.split(" ")
    line_count = 0
    split_input = ""
    for word in words:
        line_count += 1
        line_count += len(word)
        if line_count > line_length:
            split_input += "\n"
            line_count = len(word) + 1
            split_input += word
            split_input += " "
        else:
            split_input += word
            split_input += " "
    
    return split_input
```

```python
encuestados_rosario = pd.read_excel("Rosario F.xlsx")
# Eliminamos los datos que están por fuera de nuestro mapa de la ciudad de Rosario
encuestados_rosario = encuestados_rosario[encuestados_rosario.Coord1__Latitude < -32.85]
encuestados_rosario = encuestados_rosario[encuestados_rosario.Coord1__Longitude < -60.5]
encuestados_rosario = encuestados_rosario[encuestados_rosario.Coord1__Longitude > -60.78]

geo_encuestados = gpd.GeoDataFrame(
    encuestados_rosario, geometry=gpd.points_from_xy(encuestados_rosario.Coord1__Longitude,encuestados_rosario.Coord1__Latitude, crs="EPSG:4326")
)

geo_encuestados = geo_encuestados.drop(["interview__key", "Coord1__Latitude", "Coord1__Longitude", "RADIO"], axis='columns')
```

```python
pd.set_option("display.max_colwidth", None) # Muestra todas las celdas completas
barrios_df = pd.read_csv("barrios_json.csv", encoding="cp1252", dtype={"GEOJSON": str})
#print(barrios_df["GEOJSON"])
def convertir_a_geometria(string_geojson):
    # Cargar el string como objeto JSON
    geojson_obj = json.loads(string_geojson)
    # Extraer la geometría y convertirla a objeto Shapely
    return shape(geojson_obj['geometry'])
barrios_df["geometry"] = barrios_df["GEOJSON"].apply(convertir_a_geometria)
barrios_df = barrios_df.drop(columns=["MSLINK", "GEOJSON", "OBSERVACIO"])
mapa_barrios = gpd.GeoDataFrame(barrios_df, geometry='geometry')
mapa_barrios = mapa_barrios.set_crs("EPSG:4326", 4326)
```

```python
# Mapa con los nombres de los barrios y los lugares donde es encuestó

fig, ax = plt.subplots(figsize=(16,16))
mapa_barrios.plot(ax=ax, color="None", edgecolor="black")
etiquetas = []
for geometria, barrio in zip(mapa_barrios.geometry, mapa_barrios['BARRIO']):
    geometria_x, geometria_y = geometria.centroid.x, geometria.centroid.y
    area = geometria.area
    geometria_x, geometria_y = geometria.representative_point().x, geometria.representative_point().y
    barrio = split_str_into_lines(barrio, 13)
    
    if area * 10000 > 2:
        etiquetas.append(ax.text(geometria_x, geometria_y, barrio, fontsize=14, ha='center', color='red'))
    elif area * 10000 > 1:
        etiquetas.append(ax.text(geometria_x, geometria_y, barrio, fontsize=12, ha='center', color='red'))
    else:
        etiquetas.append(ax.text(geometria_x, geometria_y, barrio, fontsize=7, ha='center', color='red'))
adjust_text(etiquetas, ax=ax, avoid_self=False, arrowprops=dict(arrowstyle='->', color='red'))
ax.set_title("Mapa de Rosario", fontsize=16)
ax.set_xlabel("Longitud")
ax.set_ylabel("Latitud")

geo_encuestados.plot(ax=ax, marker="o", markersize=5, color="blue")
plt.show()
```

```python
mapa_barrios["N_encuestados"] = 0
for fila, barrio in mapa_barrios.iterrows():
    n = geo_encuestados[geo_encuestados.geometry.within(barrio.geometry)].shape[0]
    mapa_barrios.at[fila, "N_encuestados"] = n

fig, ax = plt.subplots(figsize=(8,8))
mapa_barrios.plot(
    ax = ax,
    edgecolor = "black",
    column = "N_encuestados",
    legend = True,
    legend_kwds = {"label": "Encuestados por barrio", "orientation": "vertical"},
    cmap = "Blues"
)

plt.show()
```

```python
# Fusionamos los individuos encuestados con los datos del barrio


mapa_barrios["individuos"] = mapa_barrios.apply(
    lambda row: geo_encuestados[geo_encuestados.geometry.within(row.geometry)], axis=1
)

# Crear una función para contar los valores de la columna 'columna_interes' en los individuos de cada barrio
def contar_valores(sub_gdf, columna_interes):
    if sub_gdf.empty:
        return {}
    # Contar ocurrencias de cada valor en la columna
    return sub_gdf[columna_interes].value_counts().to_dict()

# Aplicar la función a cada fila del GeoDataFrame de barrios
mapa_barrios['conteo_valores'] = mapa_barrios['individuos'].apply(
    lambda sub_gdf: contar_valores(sub_gdf, 'INT DE VOTO X ESPACIO')
)
# Expandir los conteos en columnas separadas
conteos_expandido = mapa_barrios['conteo_valores'].apply(pd.Series).fillna(0)
# Agregar estas columnas al GeoDataFrame de barrios
mapa_barrios = pd.concat([mapa_barrios, conteos_expandido], axis=1)

mapa_barrios = mapa_barrios.loc[:,~mapa_barrios.columns.duplicated()]

print(mapa_barrios.columns)
```

```python
mapa_barrios["NINGUNO/NOSABE_PCT"] = (mapa_barrios["NINGUNO/NOSABE"] * 100) / mapa_barrios["N_encuestados"]
mapa_barrios["OTROS_PCT"] = (mapa_barrios["OTROS"] * 100) / mapa_barrios["N_encuestados"]
mapa_barrios["LIBERTARIOS_PCT"] = (mapa_barrios["LIBERTARIOS"] * 100) / mapa_barrios["N_encuestados"]
mapa_barrios["FIT_PCT"] = (mapa_barrios["FIT"] * 100) / mapa_barrios["N_encuestados"]
mapa_barrios["PNK_PCT"] = (mapa_barrios["PNK"] * 100) / mapa_barrios["N_encuestados"]
mapa_barrios["JxC_PCT"] = (mapa_barrios["JxC"] * 100) / mapa_barrios["N_encuestados"]
mapa_barrios["FdT_PCT"] = (mapa_barrios["FdT"] * 100) / mapa_barrios["N_encuestados"]

norm = Normalize(vmin=0, vmax=100)
```

```python
def dibujar_intencion_voto_por_barrio(columna, cmap, label, titulo, export = False, show = True):
    
    fig, ax = plt.subplots(figsize=(8,8))
    mapa_barrios.plot(
        ax = ax,
        edgecolor = "black",
        column = columna,
        legend = True,
        legend_kwds = {"label": label, "orientation": "vertical"},
        cmap = cmap,
        norm = norm
    )
    ax.set_title(f"Rosario ({titulo})", fontsize=16)
    if export == True: 
        plt.savefig(titulo)
    if show == True:
        plt.show()

#Votos NINGUNO/NOSABE
dibujar_intencion_voto_por_barrio("NINGUNO/NOSABE_PCT", "Greys", "Votos en blanco o indecisos por barrio (en porcentaje)", "votos indecisos o en blanco")
```

```python
# Votos a otros
dibujar_intencion_voto_por_barrio("OTROS_PCT", "Greys", "Votos a otros por barrio (en porcentaje)", "votos a partidos minoritarios")
```
